# News Dashboard – Exploratory Data Analysis

Use this notebook for ad-hoc exploration of the DuckDB database.

```python
import sys
sys.path.insert(0, '..')

import duckdb
import pandas as pd
import plotly.express as px

conn = duckdb.connect('../data/processed/news.duckdb')
df = conn.execute('SELECT * FROM articles ORDER BY published_utc DESC LIMIT 1000').df()
print(f'Loaded {len(df)} articles')
df.head()
```

```python
# Category distribution
px.histogram(df, x='category', color='sentiment', barmode='group', title='Articles by Category & Sentiment')
```

```python
# Confidence distribution by category
px.box(df[df['category_conf'] > 0], x='category', y='category_conf', title='Classifier Confidence by Category')
```

```python
# Daily volume
df['date'] = pd.to_datetime(df['published_utc']).dt.date
daily = df.groupby(['date', 'category']).size().reset_index(name='count')
px.line(daily, x='date', y='count', color='category', title='Daily Article Volume by Category')
```
